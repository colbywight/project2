# Import Packages

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# Read Data

```python
data = pd.read_csv("crime-housing-austin-2015.csv")
data
```

# Get rid of percentage and dollar signs for Median House Hold Income and Median Home value.  Separate data into different crimes.

```python
data['Medianhouseholdincome'] = data['Medianhouseholdincome'].str.replace('$', '').astype('float')
data['Medianhomevalue'] = data['Medianhomevalue'].str.replace('$', '').astype('float')
theft = data[data.Highest_NIBRS_UCR_Offense_Description == 'Theft']
robbery = data[data.Highest_NIBRS_UCR_Offense_Description == 'Robbery']
burglary = data[data.Highest_NIBRS_UCR_Offense_Description == 'Burglary']
auto = data[data.Highest_NIBRS_UCR_Offense_Description == 'Auto Theft']
assault = data[data.Highest_NIBRS_UCR_Offense_Description == 'Agg Assault']
rape = data[data.Highest_NIBRS_UCR_Offense_Description == 'Rape']
murder = data[data.Highest_NIBRS_UCR_Offense_Description == 'Murder']
```
