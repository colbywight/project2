# T-tests and Pearson R values

```python
sns.distplot(theft.Medianhouseholdincome, label='Theft', hist=False)
sns.distplot(burglary.Medianhouseholdincome, label='Burglary', hist=False)
sns.distplot(auto.Medianhouseholdincome, label='Auto Theft', hist=False)
sns.distplot(assault.Medianhouseholdincome, label='Agg Assault', hist=False)
sns.distplot(robbery.Medianhouseholdincome, label='Robbery', hist=False)
sns.distplot(rape.Medianhouseholdincome, label='Rape', hist=False)
sns.distplot(murder.Medianhouseholdincome, label='Murder', hist=False)
plt.title('All crimes combined')

plt.figure()
sns.distplot(theft.Medianhouseholdincome.dropna(), label='Theft', hist=False)
sns.distplot(auto.Medianhouseholdincome.dropna(), label='Auto Theft', hist=False)
plt.title('Theft and Auto Theft')

plt.figure()
sns.distplot(theft.Medianhouseholdincome, label='Theft', hist=False)
sns.distplot(assault.Medianhouseholdincome, label='Agg Assault', hist=False)
plt.title('Theft and Agg Assault')

plt.figure()
sns.distplot(theft.Medianhouseholdincome, label='Theft', hist=False)
sns.distplot(robbery.Medianhouseholdincome, label='Robbery', hist=False)
plt.title('Theft and Robbery')

plt.figure()
sns.distplot(theft.Medianhouseholdincome, label='Theft', hist=False)
sns.distplot(burglary.Medianhouseholdincome, label='Burglary', hist=False)
plt.title('Theft and Burglary')

plt.figure()
sns.distplot(theft.Medianhouseholdincome, label='Theft', hist=False)
sns.distplot(rape.Medianhouseholdincome, label='Rape', hist=False)
plt.title('Theft and Rape')

plt.figure()
sns.distplot(theft.Medianhouseholdincome, label='Theft', hist=False)
sns.distplot(murder.Medianhouseholdincome, label='Murder', hist=False)
plt.title('Theft and Murder')

plt.figure()
sns.distplot(robbery.Medianhouseholdincome, label='Robbery', hist=False)
sns.distplot(robbery.Medianhouseholdincome, label='Robbery', hist=False)
plt.title('Theft and Robbery')

theft1 = theft.Medianhouseholdincome.dropna()
auto1 = auto.Medianhouseholdincome.dropna()
burglary1 = burglary.Medianhouseholdincome.dropna()
assault1 = assault.Medianhouseholdincome.dropna()
robbery1 = robbery.Medianhouseholdincome.dropna()
rape1 = rape.Medianhouseholdincome.dropna()
murder1 = murder.Medianhouseholdincome.dropna()

```

## T tests (difference between groups, probability of receiving this difference)

```python
# display(stats.ttest_ind(df[df.State == 'UT'].Score, df[df.State == 'AL'].Score))
display("-------Theft-------")
display(stats.ttest_ind(theft1, burglary1))
display(stats.ttest_ind(theft1, auto1))
display(stats.ttest_ind(theft1, assault1))
display(stats.ttest_ind(theft1, robbery1))
display(stats.ttest_ind(theft1, rape1))
display(stats.ttest_ind(theft1, murder1))
display("-------Burglary-------")
display(stats.ttest_ind(burglary1, auto1))
display(stats.ttest_ind(burglary1, assault1))
display(stats.ttest_ind(burglary1, robbery1))
display(stats.ttest_ind(burglary1, rape1))
display(stats.ttest_ind(burglary1, murder1))
display("-------Auto Theft-------")
display(stats.ttest_ind(auto1, assault1))
display(stats.ttest_ind(auto1, robbery1))
display(stats.ttest_ind(auto1, rape1))
display(stats.ttest_ind(auto1, murder1))
display("-------Agg Assault-------")
display(stats.ttest_ind(assault1, robbery1))
display(stats.ttest_ind(assault1, rape1))
display(stats.ttest_ind(assault1, murder1))
display("-------Robbery-------")
display(stats.ttest_ind(robbery1, rape1))
display(stats.ttest_ind(robbery1, murder1))
display("-------Rape and Murder-------")
display(stats.ttest_ind(rape1, murder1))
```

```python
theftcorr = theft.groupby("Medianhomevalue")
burglarycorr = burglary.groupby("Medianhomevalue")
autocorr = auto.groupby("Medianhomevalue")
assaultcorr = assault.groupby("Medianhomevalue")
robberycorr = robbery.groupby("Medianhomevalue")
rapecorr = rape.groupby("Medianhomevalue")
murdercorr = murder.groupby("Medianhomevalue")
```

```python
theftcorr = theftcorr.agg({'Highest_NIBRS_UCR_Offense_Description':'size', 'Medianhomevalue':'median', 'Medianhouseholdincome':'median'})
burglarycorr = burglarycorr.agg({'Highest_NIBRS_UCR_Offense_Description':'size', 'Medianhomevalue':'median', 'Medianhouseholdincome':'median'})
autocorr = autocorr.agg({'Highest_NIBRS_UCR_Offense_Description':'size', 'Medianhomevalue':'median', 'Medianhouseholdincome':'median'})
assaultcorr = assaultcorr.agg({'Highest_NIBRS_UCR_Offense_Description':'size', 'Medianhomevalue':'median', 'Medianhouseholdincome':'median'})
robberycorr = robberycorr.agg({'Highest_NIBRS_UCR_Offense_Description':'size', 'Medianhomevalue':'median', 'Medianhouseholdincome':'median'})
rapecorr = rapecorr.agg({'Highest_NIBRS_UCR_Offense_Description':'size', 'Medianhomevalue':'median', 'Medianhouseholdincome':'median'})
murdercorr = murdercorr.agg({'Highest_NIBRS_UCR_Offense_Description':'size', 'Medianhomevalue':'median', 'Medianhouseholdincome':'median'})
```

### R values, two tail probabilities

```python
display(stats.pearsonr(theftcorr.Medianhomevalue, theftcorr.Highest_NIBRS_UCR_Offense_Description))
display(stats.pearsonr(burglarycorr.Medianhomevalue, burglarycorr.Highest_NIBRS_UCR_Offense_Description))
display(stats.pearsonr(autocorr.Medianhomevalue, autocorr.Highest_NIBRS_UCR_Offense_Description))
display(stats.pearsonr(assaultcorr.Medianhomevalue, assaultcorr.Highest_NIBRS_UCR_Offense_Description))
display(stats.pearsonr(robberycorr.Medianhomevalue, robberycorr.Highest_NIBRS_UCR_Offense_Description))
display(stats.pearsonr(rapecorr.Medianhomevalue, rapecorr.Highest_NIBRS_UCR_Offense_Description))
display(stats.pearsonr(murdercorr.Medianhomevalue, murdercorr.Highest_NIBRS_UCR_Offense_Description))
```
